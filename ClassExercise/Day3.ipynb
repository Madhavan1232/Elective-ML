```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
    {
        "id" : [1, pd.NA, 3, 4],
        "name" : ["Alice", pd.NA, "Charlie", "David"]
    }
)
df2 = pd.DataFrame(
    {
        "id" : [3, 4, 5, pd.NA],
        "name" : ["Charlie", "David", pd.NA, "Frank"]
    }
)

df3 = pd.DataFrame(
    {
        "age" : [25, 30, 35, 40],
        "city" : ["New York", pd.NA, "Chicago", "Houston"]
    }
)

df4 = pd.DataFrame(
    {
        "age" : [35, 40, 45, 50],
        "city" : ["Chicago", "Houston", "Phoenix", "Philadelphia"]
    }
)

res = pd.concat([df1, df2], ignore_index=True)
res1 = pd.concat([df3, df4], ignore_index=True)
final = pd.merge(res, res1, how='outer' , left_index=True, right_index=True)
print(final.dropna(how='all'))
final.describe()
```
