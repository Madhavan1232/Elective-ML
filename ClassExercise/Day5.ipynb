```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = ['tips' , 'iris' , 'titanic' , 'flights' , 'diamonds' , 'penguins']
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
diamonds = sns.load_dataset('diamonds')

for data in dataset:
    display(eval(data).head())
```

```python
sns.scatterplot(data = iris, x = 'sepal_length' , y  = 'sepal_width' , color = 'red' )
plt.show() 
```

```python
sns.lineplot(data = flights, x = 'month' , y = 'passengers' , hue = 'year' )
plt.show()
```

```python
sns.countplot(data = penguins, x = 'bill_depth_mm' , hue = 'sex' )
plt.show()
```

```python
sns.boxplot(data = titanic , x = 'class' , y = 'age' , hue = 'sex' )
plt.show()
```
