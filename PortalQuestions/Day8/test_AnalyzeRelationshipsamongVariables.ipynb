```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the data
df = pd.read_excel('Data/DA314_S8_EmployeePerformance_Data_Concept.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
correlation = df['avg_training_score'].corr(df['performance_score'])
ax.scatter(df['avg_training_score'], df['performance_score'], alpha=0.6, s=50, color='red')
ax.set_xlabel('avg_training_score')
ax.set_ylabel('performance_score')
ax.set_title(f'Correlation: {correlation:.2f}')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Task 2: Heatmap of Numeric Features
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, ax=ax, cbar_kws={'label': 'Correlation'})
ax.set_title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```

```python
fig = sns.pairplot(numeric_df)
plt.tight_layout()
plt.show()
```
