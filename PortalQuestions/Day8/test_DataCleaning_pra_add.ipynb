```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S8_EmployeeAttrition_Data_Practice.csv')
```

```python
# Detect outliers in PerformanceRating using Boxplot and IQR

# Calculate IQR and outlier thresholds
Q1 = df['PerformanceRating'].quantile(0.25)
Q3 = df['PerformanceRating'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR



# Identify outliers
outliers = df[(df['PerformanceRating'] < lower_bound) | (df['PerformanceRating'] > upper_bound)]



# Create boxplot visualization
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['PerformanceRating'], color='skyblue')
plt.title('Outlier Detection in Performance Rating', fontsize=14, fontweight='bold')
plt.ylabel('PerformanceRating', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Additional statistics
print("\nPerformanceRating Statistics:")
print("=" * 60)
print(df['PerformanceRating'].describe())
```
