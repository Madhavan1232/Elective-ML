```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S5_GlamNest_Data_Practice.csv')
```

```python
fig , aix = plt.subplots(1,2, figsize=(15,5) , tight_layout=True)
sns.histplot(data = df , x = 'final_amount' , kde = True , color = 'skyblue' , bins = 13 , ax=aix[0])
aix[0].set_title("Service Amount Distribution")
aix[0].set_ylabel("Count")
aix[0].grid(linestyle='dotted')

sns.boxplot(data = df , x = 'final_amount' , color = 'lightcoral' , ax=aix[1])
aix[1].set_title("Service Amount Spread")
plt.xlabel("Final Amount")
aix[1].grid(linestyle='dotted')
```

```python
fig , axis = plt.subplots(1 , 2 , figsize=(15,5) , tight_layout=True)
sns.histplot(data = df , x = 'service_rating' , kde = True , color = 'lightcoral' , bins = 13 , ax=axis[0])
axis[0].set_title("Service Rating Distribution")
axis[0].set_ylabel("Count")
axis[0].grid(linestyle='dotted')

sns.boxplot(data = df , x = 'service_rating' , color = 'navajowhite' , ax=axis[1])
axis[1].set_title("Service Rating Spread")
plt.xlabel("Service Rating")
axis[1].grid(linestyle='dotted')
```
