```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S5_GlamNest_Data_Practice.csv')
```

```python
data = df[['price' , 'discount%' , 'discounted_amount(INR)' , 'final_amount' , 'service_rating']].corr()
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
