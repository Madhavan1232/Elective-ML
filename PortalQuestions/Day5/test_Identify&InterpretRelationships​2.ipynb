```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S5_OrderDetails_Data_Concept.csv')
```

```python
data = df[['avg_cost_for_two' , 'rating']].corr()
sns.heatmap(data, annot = True , cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.regplot(x = 'avg_cost_for_two' , y = 'rating' , data = df , color = 'orange' , line_kws={'linewidth':2 , 'color':'red' , 'linestyle':'--'})
plt.title('cost vs Customer Rating with Trend Line')
plt.xlabel('Average Cost for Two')
plt.grid()
plt.ylabel('Customer Rating')
plt.show()
```
