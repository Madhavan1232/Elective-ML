```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S4_GlamNest_Data_Practice.csv')
```

```python
plt.hist(df['final_amount'] , bins=12 , edgecolor='black' , color='skyblue' , alpha=0.7)
plt.title("Distribution of Final Amount")
plt.xlabel("Final Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.kdeplot(data=df, x='service_rating', fill=True, color='lightgreen', alpha=0.5 , edgecolor='green')
plt.title("KDE Plot of Service Ratings")
plt.xlabel("Rating")
plt.ylabel("Density")
plt.show()
```

```python
sns.boxplot(x = 'category' , y = 'final_amount' , data = df)
plt.xticks(rotation=45)
plt.show()
```
