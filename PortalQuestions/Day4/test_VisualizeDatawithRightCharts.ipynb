```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('../../data/DA314_S4_OrderDetails_Data_Concept.csv')
```

```python
plt.figure(figsize=(10,6))
plt.pie(data['payment_method'].value_counts(), labels=data['payment_method'].value_counts().index, autopct='%1.1f%%')
plt.title('Payment Methods Share')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sorted_df = data["city_user"].value_counts().sort_values()
plt.barh(sorted_df.index, sorted_df)
plt.show()
```

```python
data['order_date'] = pd.to_datetime(data['order_date'], errors='coerce', infer_datetime_format=True)
monthly_orders = data.groupby(data['order_date'].dt.to_period('M') ).size()

plt.figure(figsize=(10, 6))
plt.plot(monthly_orders.index.astype(str), monthly_orders.values,  linestyle='-', marker='o', linewidth=2, markersize=6)
plt.title('Total Orders by Month')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.xticks(rotation=65)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_rating = data['rating'].value_counts().sort_index()
plt.scatter(x = "rating", y = "total_amount", data = data, alpha=0.9 , color='green')
plt.title('Total Amount vs Product Rating')
plt.xlabel('Product Rating')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

```python
summary = data.groupby(["city_restaurant", "payment_method"])['total_amount'].sum().reset_index()
pivot_table = summary.pivot(index='city_restaurant', columns='payment_method', values='total_amount').fillna(0)

pivot_table.plot(kind='bar', stacked=True)
plt.title('Total Amount by City and Payment Method')
plt.xlabel('City') 
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```
