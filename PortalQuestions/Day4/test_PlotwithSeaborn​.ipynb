```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('../../data/DA314_S4_OrderDetails_Data_Concept.csv')
```

```python
sns.violinplot(x = "city_restaurant", y = "delivery_time_min", data = df , color='wheat')
plt.title("Delivery Time Distribution by City Restaurant")
plt.xlabel("City Restaurant")
plt.ylabel("Delivery Time (min)")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df , x_vars=['total_amount', 'delivery_time_min' , "rating"], y_vars=['total_amount', 'delivery_time_min' , "rating"] , plot_kws={'color':'orange'})
plt.show()
```

```python
fig , axes = plt.subplots(3 , 1)

sns.histplot(df['total_amount'] , kde = True , ax = axes[0] , bins= 12)
axes[0].set_title("Distribution of Total Amount")
axes[0].set_xlabel("Total Amount")
axes[0].set_ylabel("count")

sns.histplot(df['rating'] , kde=True , ax = axes[1] , bins= 12)
axes[1].set_title("Distribution of Ratings")
axes[1].set_xlabel("Ratings")
axes[1].set_ylabel("count")

sns.histplot(df['delivery_time_min'] , kde=True , ax = axes[2] , bins= 12)
axes[2].set_title("Distribution of Delivery Time")
axes[2].set_xlabel("Delivery Time")
axes[2].set_ylabel("count")

plt.tight_layout()
```

```python
fig , axes = plt.subplots(3 ,1 , figsize=(12,5))

sns.histplot(data=df['total_amount'] , kde=True , bins=12 , ax = axes[0])
sns.histplot(data=df["rating"] , kde=True , bins=12 , ax = axes[1])
sns.histplot(data=df["delivery_time_min"] , kde=True , bins=12 , ax = axes[2])
plt.show()
```
