```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S4_GlamNest_Data_Practice.csv')
```

```python
completed_apo = df[df['status'] == 'Completed']
service_count = completed_apo['category'].value_counts()
service_count.plot(kind='bar', color='skyblue')
plt.title('Completed Appointments by Service Category')
plt.xlabel('Service Category')
plt.ylabel('Count')
plt.show()
```

```python
df['appointment_date'] = pd.to_datetime(df['appointment_date'], format='%d/%m/%Y')
df['year_month'] = df['appointment_date'].dt.month_name()
monthly_avg_revenue = df.groupby('year_month')['final_amount'].mean().reset_index()
monthly_avg_revenue['year_month'] = monthly_avg_revenue['year_month'].astype(str)
plt.plot(monthly_avg_revenue['year_month'], monthly_avg_revenue['final_amount'], marker='o')
plt.title('Average Monthly Revenue')
plt.xlabel("Month")
plt.ylabel("Avg revenue")
plt.xticks(rotation = 65)
plt.show()
```

```python
grouped_df = df.groupby(['city' , 'status']).size().reset_index(name = 'counts')
sns.barplot(data=grouped_df , x = 'city' , y = 'counts' , hue = 'status')
plt.title("Appointments by status and city")
plt.xticks(rotation = 55)
plt.show()
```

```python
grouped_payment = df.groupby(['city', 'payment_method']).size().unstack()
grouped_payment.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Payment Method Distribution Across Cities')
plt.xlabel('City')
plt.ylabel('Count of Appointments')
plt.xticks(rotation=45)
plt.legend(title='Payment Method')
plt.show()
```

```python
plt.pie(df['category'].value_counts(),  autopct='%1.1f%%', startangle=140 , labels=df['category'].value_counts().index)
plt.title('Service Category Distribution')
```

```python
sns.scatterplot(data = df , x ='price' , y = 'rating' , hue = 'category')
plt.title("Service Price vs. Customer Rating", fontsize=13)
plt.xlabel("Service Price(INR)")
plt.ylabel("Rating")
plt.legend(title="category")
plt.show()
```
