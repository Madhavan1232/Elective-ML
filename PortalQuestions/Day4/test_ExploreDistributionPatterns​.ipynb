```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('../../data/DA314_S4_OrderDetails_Data_Concept.csv')
```

```python
freq = df['total_amount'].value_counts()
plt.hist(df['total_amount'], bins=8 , color="skyblue", edgecolor='black')
plt.title('Histogram of Total Amount') 
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(df["total_amount"], fill=True , color='orange')
plt.title('KDE Plot of Order Amount')
plt.xlabel('Order Amount')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x = df['city_restaurant'] , y = df['delivery_time_min'])
plt.title('Box Plot of Delivery Time by City Restaurant')
plt.xlabel('City Restaurant')
plt.ylabel('Delivery Time (minutes)')
plt.xticks(rotation=45)
plt.show()
```
