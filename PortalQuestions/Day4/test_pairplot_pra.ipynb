```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data/DA314_S4_GlamNest_Data_Practice.csv')
```

```python
sns.violinplot(data = df , x = 'gender' , y = 'final_amount' , color = 'skyblue')
```

```python
sns.pairplot(data = df , vars=['price' , 'final_amount' , 'rating'])
plt.show()
```

```python
mumbai_df = df[df['city'] == 'Mumbai']
delhi_df = df[df['city'] == 'Delhi']
fig , aix = plt.subplots(1 , 2 , figsize = (12 , 5) )
sns.histplot(data = mumbai_df , x = 'final_amount' , ax = aix[0] )
sns.histplot(data = delhi_df , x = 'final_amount'  , ax = aix[1] )
```
